```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    """Computes the sigmoid function and its derivative.

    Args:
        x: A NumPy array of input values.

    Returns:
        A tuple containing the sigmoid and its derivative (both NumPy arrays).
    """
    s = 1 / (1 + np.exp(-x))
    ds = s * (1 - s)
    return s, ds

# Generate x values for plotting
x = np.arange(-6, 6, 0.01)

# Calculate sigmoid and its derivative
sigmoid_x, derivative_sigmoid_x = sigmoid(x)

# Create the plot figure and axes
fig, ax = plt.subplots(figsize=(9, 5))

# Customize the axes (centering and removing spines)
ax.spines['left'].set_position('center')
ax.spines['right'].set_visible(False)  # More concise way to remove spine
ax.spines['top'].set_visible(False)    # More concise way to remove spine
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Plot the sigmoid function
ax.plot(x, sigmoid_x, color="#307EC7", linewidth=3, label="Sigmoid Function")

# Plot the derivative of the sigmoid function
ax.plot(x, derivative_sigmoid_x, color="#9621E2", linewidth=3, label="Sigmoid Derivative")

# Add a legend (without a frame)
ax.legend(loc="upper right", frameon=False)

# Add axis labels for clarity (important for coursework!)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")

# Add a title to the plot
ax.set_title("Sigmoid Function and its Derivative")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def ReLU(x):
    """Computes the ReLU function and its derivative.

    Args:
        x: A NumPy array of input values.

    Returns:
        A tuple containing the ReLU output and its derivative.
    """
    t = x * (x > 0) 
    dt = 1 * (x > 0)  # Forward pass for ReLU: y = max(0, x)
# Backward pass for ReLU: derivative is 1 for x > 0, otherwise 0

    return t, dt

# Generate x values for plotting
x = np.linspace(-4, 4, 400)  # Use linspace for smoother curves

# Calculate ReLU and its derivative
relu_x, derivative_relu_x = ReLU(x)

# Create the plot figure and axes
fig, ax = plt.subplots(figsize=(10, 6))  # Slightly larger figure

# Customize the axes (centering and removing spines)
ax.spines['left'].set_position('zero')  # Center at zero
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Plot the ReLU function
ax.plot(x, relu_x, color="#307EC7", linewidth=2, label="ReLU(x)")

# Plot the derivative of the ReLU function
ax.plot(x, derivative_relu_x, color="#9621E2", linewidth=2, label="ReLU'(x)")

# Add a legend
ax.legend(loc="upper left", frameon=False)

# Add axis labels and title
ax.set_xlabel("x", loc='right')  # Position labels at the end of the axes
ax.set_ylabel("y", loc='top', rotation=0)

ax.set_title("ReLU Function and its Derivative", fontsize=14, fontweight='bold')

# Add grid for better visualization (optional)
ax.grid(True, linestyle='--', alpha=0.5)

# Improve ticks for better readability
ax.set_xticks(np.arange(-4, 5, 1))
ax.set_yticks(np.arange(0, 5, 1))

# Set limits for y-axis to improve visualization of derivative
ax.set_ylim(-0.5, 4)  # Adjusted y-axis limits

# Add arrows to axes
ax.plot((1), (0), ls="", marker=">", ms=10, color="k",
        transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0), (1), ls="", marker="^", ms=10, color="k",
        transform=ax.get_xaxis_transform(), clip_on=False)

# Show the plot
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python

import numpy as np

def sigmoid(x):
    """Computes the sigmoid function."""
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    """Computes the derivative of the sigmoid function."""
    s = sigmoid(x)
    return s * (1 - s)

def relu(x):
    """Computes the ReLU function."""
    return np.maximum(0, x) #More robust

def relu_derivative(x):
    """Computes the derivative of the ReLU function."""
    return np.where(x > 0, 1, 0) #More robust


# Example usage with different input types
test_values_scalar = 0.4
test_values_array = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])
test_values_int = np.array([-2, -1, 0, 1, 2])

print("--- Scalar Input ---")
print("Sigmoid(", test_values_scalar, "):", sigmoid(test_values_scalar))
print("Sigmoid Derivative(", test_values_scalar, "):", sigmoid_derivative(test_values_scalar))
print("ReLU(", test_values_scalar, "):", relu(test_values_scalar))
print("ReLU Derivative(", test_values_scalar, "):", relu_derivative(test_values_scalar))

print("--- Array Input ---")
print("Sigmoid(", test_values_array, "):", sigmoid(test_values_array))
print("Sigmoid Derivative(", test_values_array, "):", sigmoid_derivative(test_values_array))
print("ReLU(", test_values_array, "):", relu(test_values_array))
print("ReLU Derivative(", test_values_array, "):", relu_derivative(test_values_array))

print("--- Int Array Input ---")
print("ReLU Derivative(", test_values_int, "):", relu_derivative(test_values_int))
```
